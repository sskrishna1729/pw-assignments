### Seaborn is a popular data visualization library in Python that provides a high-level interface for creating informative and attractive statistical graphics. You can create a wide variety of plots using Seaborn to explore and communicate data effectively. Here are five types of plots you can create with Seaborn, along with their common uses:

## Scatter Plot:

### Use: Scatter plots are used to visualize the relationship between two continuous variables. They help you understand patterns, correlations, and outliers in your data.
## Histogram:

## Use: Histograms display the distribution of a single variable by dividing it into bins and showing the frequency of data points within each bin. They are useful for understanding the data's central tendency and spread.
## Bar Plot:

### Use: Bar plots are used to represent categorical data. They display the distribution or comparison of categories by plotting their frequencies or aggregated values. They are helpful for comparing different categories or groups.
## Box Plot (Box-and-Whisker Plot):

### Use: Box plots summarize the distribution of a continuous variable, showing the median, quartiles, and potential outliers. They are particularly useful for visualizing the spread and skewness of data and identifying potential outliers.
## Heatmap:

### Use: Heatmaps are useful for visualizing the correlation or relationships between multiple variables in a matrix format. They use colors to represent the strength and direction of relationships, making them ideal for exploring complex datasets.

# Question-2

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Use Seaborn to create the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

# Add labels and a legend
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Data: Signal vs. Timepoint")
plt.legend(title="Event", loc="best")

# Show the plot
plt.show()
```

# Question-3

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots (one for age and one for fare)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Create box plots for 'age' and 'fare' by 'pclass'
sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
axes[0].set_title('Age by Pclass')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Age')

sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])
axes[1].set_title('Fare by Pclass')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Fare')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

# Question-4

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram with 'price' and 'cut' as the hue
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Use Seaborn to create the histogram
sns.histplot(data=diamonds, x="price", hue="cut", bins=30, kde=True, palette="viridis")

# Add labels and a title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Price Histogram by Cut")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Reshape the data for a heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Passenger Counts by Month and Year')
plt.show()
```
