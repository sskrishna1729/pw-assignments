```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title('Question2_ScatterPlot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

# Question-3

### The subplot() function in Matplotlib is used to create multiple plots within the same figure, allowing you to display different visualizations side by side or in a grid layout. It's particularly useful when you want to compare and visualize multiple datasets simultaneously.

```python
# Data for the four line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Create a figure with a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1)
plt.title("Line Plot 1")

# Subplot 2
plt.subplot(2, 2, 2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2)
plt.title("Line Plot 2")

# Subplot 3
plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3)
plt.title("Line Plot 3")

# Subplot 4
plt.subplot(2, 2, 4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4)
plt.title("Line Plot 4")

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

# Question-4

### A bar plot, also known as a bar chart or bar graph, is a data visualization that uses rectangular bars to represent data. It is particularly useful for displaying and comparing the values of different categories or groups. Bar plots are typically used for categorical data, and they provide a straightforward way to visualize data distributions, comparisons, and trends.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Company Profits (Vertical Bar Plot)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit)
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.show()
```

# Question-5

### A box plot, also known as a whisker plot, is a graphical representation of a data distribution that provides information about the center, spread, and potential outliers in the data. It is commonly used to visualize the distribution of a dataset and identify key summary statistics. 

### Median: The median, also known as the second quartile (Q2), is represented by a horizontal line inside the box. It represents the middle value of the dataset when the data is ordered.

### Quartiles: The box is divided into two parts by the median, and each half is further divided into quartiles. The lower quartile (Q1) and the upper quartile (Q3) represent the 25th and 75th percentiles of the data, respectively.

### Interquartile Range (IQR): The IQR is the range between Q1 and Q3, and it contains the middle 50% of the data.

### Whiskers: The whiskers extend from the box to the minimum and maximum values within a defined range. Data points beyond the whiskers are considered potential outliers.

### Outliers: Data points outside the whiskers are typically shown as individual points or asterisks and may be considered outliers

```python
# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
fig, ax = plt.subplots()

# Create a box plot
boxplot_data = [box1, box2]
ax.boxplot(boxplot_data)

# Customize the plot
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')

# Show the plot
plt.show()
```
