```python
from bs4 import BeautifulSoup
from selenium import webdriver

# Replace with the URL of the YouTube channel's video page
url = "https://www.youtube.com/@PW-Foundation/videos"

# Set up a headless browser (you may need to download the appropriate web driver)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(executable_path='path_to_chromedriver', options=options)

# Load the page
driver.get(url)

# Wait for the page to load (you might need to adjust the sleep time)
import time
time.sleep(5)

# Extract the page source
page_source = driver.page_source

# Close the browser
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find the video URLs
video_urls = []
for link in soup.find_all('a', {'id': 'video-title'}):
    video_url = "https://www.youtube.com" + link.get('href')
    video_urls.append(video_url)

# Extract the first five video URLs
first_five_video_urls = video_urls[:5]

for i, url in enumerate(first_five_video_urls, start=1):
    print(f"Video {i}: {url}")
```

# Question-4

```python
from bs4 import BeautifulSoup
from selenium import webdriver

# Replace with the URL of the YouTube channel's video page
url = "https://www.youtube.com/@PW-Foundation/videos"

# Set up a headless browser (you may need to download the appropriate web driver)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(executable_path='path_to_chromedriver', options=options)

# Load the page
driver.get(url)

# Wait for the page to load (you might need to adjust the sleep time)
import time
time.sleep(5)

# Extract the page source
page_source = driver.page_source

# Close the browser
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find the video view counts
view_counts = []
for count in soup.find_all('span', {'class': 'style-scope ytd-video-meta-block'}):
    if "views" in count.text:
        view_count = count.text.split()[0]
        view_counts.append(view_count)

# Extract the view counts of the first five videos
first_five_view_counts = view_counts[:5]

for i, count in enumerate(first_five_view_counts, start=1):
    print(f"Video {i} View Count: {count}")
```

# Question-5

```python
from bs4 import BeautifulSoup
from selenium import webdriver

# Replace with the URL of the YouTube channel's video page
url = "https://www.youtube.com/@PW-Foundation/videos"

# Set up a headless browser (you may need to download the appropriate web driver)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(executable_path='path_to_chromedriver', options=options)

# Load the page
driver.get(url)

# Wait for the page to load (you might need to adjust the sleep time)
import time
time.sleep(5)

# Extract the page source
page_source = driver.page_source

# Close the browser
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find the video publication times
publication_times = []
for time_element in soup.find_all('span', {'class': 'style-scope ytd-grid-video-renderer'}):
    time_text = time_element.text.strip()
    if time_text.startswith("Published"):
        publication_times.append(time_text)

# Extract the publication times of the first five videos
first_five_publication_times = publication_times[:5]

for i, time in enumerate(first_five_publication_times, start=1):
    print(f"Video {i} Publication Time: {time}")
```
