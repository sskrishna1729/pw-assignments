# Question-1

```python
import seaborn as sns
import plotly.express as px

# Load the "titanic" dataset using Seaborn
titanic_data = sns.load_dataset("titanic")
# Create a scatter plot using Plotly Express
scatter_plot = px.scatter(titanic_data, x="age", y="fare", title="Scatter Plot of Age vs Fare")
scatter_plot.show()
```

# Question-2

```python
import seaborn as sns
import plotly.express as px

# Load the "tips" dataset using Seaborn
tips_data = sns.load_dataset("tips")

# Create a box plot using Plotly Express
box_plot = px.box(tips_data, x="day", y="total_bill", title="Box Plot of Total Bill by Day")
box_plot.show()
```

# Question-3

```python
import seaborn as sns
import plotly.express as px

# Load the "tips" dataset using Seaborn
tips_data = sns.load_dataset("tips")

# Create a histogram using Plotly Express
histogram = px.histogram(tips_data, x="total_bill", color="day", pattern_shape="smoker", facet_col="sex",
                         title="Histogram of Total Bill by Gender, Day, and Smoking Status")
histogram.show()
```

# Question-4

```python
import seaborn as sns
import plotly.express as px

# Load the "iris" dataset using Seaborn
iris_data = sns.load_dataset("iris")

# Create a scatter matrix plot using Plotly Express
scatter_matrix = px.scatter_matrix(iris_data, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                                    color="species", title="Scatter Matrix Plot of Iris Dataset")
scatter_matrix.show()
```

# Question-5

```python
import plotly.express as px
import numpy as np

# Sample data
data = np.random.randn(1000)

# Create a density plot using Plotly Express
density_plot = px.histogram(data, nbins=30, histnorm='probability density',
                            title="Density Plot (Distribution Plot)")

density_plot.update_traces(opacity=0.75)  # Adjust the transparency for better visibility
density_plot.show()
```
